```python
#El juego funciona de tal manera que te saque preguntas de año, autor e idioma. Si la respuesta es incorrecta, te saca otra pregunta de la misma función. Si es correcta, pasas a la siguiente pregunta y terminas el juego cuando ahs acertado bien las 3. :D

import csv
import random
fin= open("input_operas.csv")
fout=open ("output_questions.csv","w")
writer=csv.writer(fout, delimiter ="\t")

lista_authors = []
lista_language = []
dic_1 = {}
#leer las lineas del csv y conseguir las partes que necesito
for line in csv.reader(fin):
  
  opera_year=int(line[0].strip())
  
  opera_name=line[1].strip()

  opera_author=line[2].strip()
  lista_authors.append(opera_author)
  
  opera_lang = line[3].strip()
  opera_language = opera_lang.replace("DE","German").replace("IT","Italian").replace("RU", "Russian").replace("FR","French")
  lista_language.append(opera_language)
  #crear diccionario 
  dic_1[opera_name] = ([opera_year,opera_author,opera_language])

#función de pregunta 1
def question_1():
  for opera in dic_1:
    operas = random.choice(list(dic_1.keys()))
    q1 = 'What year was {} composed?'.format(operas)
    print(q1)
    
    lista_distractors_years= [[dic_1[operas][0]]]
    while len(lista_distractors_years) < 4:
      year = random.sample(range(opera_year-50,opera_year+50),1)
      if year not in lista_distractors_years:          lista_distractors_years.append(year)
            
    print(random.sample(lista_distractors_years,4))
    answer = input()
    writer.writerow([q1, random.sample(lista_distractors_years,4)])
    
    def correct():
          if answer == str(dic_1[operas][0]):
                print('Great!')
          else:
              print("Oops! Keep trying")
              return question_1()
        return correct() 
#funcion de pregunta 2
def question_2():
  for opera in dic_1:
    operas = random.choice(list(dic_1.keys()))
    q2=('Next question: Who composed {}?'.format(operas))
    print(q2)
    
    lista_distractors_authors = [dic_1[operas][1]]
    while len(lista_distractors_authors) < 4:
      author= random.choice(lista_authors)
      if author not in lista_distractors_authors:
        lista_distractors_authors.append(author)
        
    print(sorted(lista_distractors_authors))
    answer = input()
    writer.writerow([q2, sorted(lista_distractors_authors)])
    
    def correct():
      if answer == dic_1[operas][1]:
        print('Great!')
      else:
          print("Oops! Keep trying")
          return question_2()
    return correct()  
#funcion de pregunta 3 
def question_3():
  for opera in dic_1:
    operas = random.choice(list(dic_1.keys()))
    q3=('Next question: In which language was {} composed?'.format(operas))
    print(q3)
  
    lista_distractors_lang = []
    while len(lista_distractors_lang) < 4:
      language = dic_1[operas][2]
      for language in lista_language:
        if language not in lista_distractors_lang:
          lista_distractors_lang.append(language)
          
    print(lista_distractors_lang)
    answer= input()
    writer.writerow([q3,lista_distractors_lang])
    
    def correct():
          if answer == dic_1[operas][2]:
                print('Great! You won!')
          else:
            print("Oops! Keep trying")
            return question_3()
        return correct()
    
   
question_1()
question_2()
question_3()

fin.close()
fout.close()
```

```python
#Getting url
import requests
def get_url(artist):
    API_ENDPOINT = 'https://www.wikidata.org/w/api.php'
    query = str(artist)
    params = { 'action' : 'wbsearchentities','format':'json','language':'en','search':query}
    r = requests.get(API_ENDPOINT, params= params)
    url = r.json()['search'][0]['url']
    return url 
```

```python
#getting image url

import requests
from bs4 import BeautifulSoup
import urllib.request

# def get_photo():
for artist in df['artist']:
    url = get_url(artist)

    html = urllib.request.urlopen('https:'+url)
    soup = BeautifulSoup(html)
    images = soup.findAll('img')
    lista = []
    final_links = []
    for image in images:
        link =image['src']
        lista.append(link)
    link_photo = lista[0]
    final_links.append(link_photo)
print(final_links)
```
