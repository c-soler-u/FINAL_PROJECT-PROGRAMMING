```python
import pandas as pd
df = pd.read_csv('final_2020.csv')
```

Creating some functions for randomizing and punctuation.

```python
import random
import textdistance 

#RANDOM GENERATION OF VARIABLE
def random_choice(dic):
    choice = random.choice(list(dic.keys()))
    return choice

#GIVING AN STATEMENT ABOUT THE ANSWER AND ADDING OR SUBSTRACTING POINTS
def punct(correct, answer,points):
        if answer.lower() == correct.lower(): #Removing capitalisation
                print('Correct!')
                points += 1
        elif textdistance.levenshtein(answer.lower(),correct.lower()) == 1: #Applying levenshtein distance. If the answer is at distance 1, take it as correct
                print('Correct!')
                points += 1
                                      
        else:
                print("Oops! The correct answer was:",correct)
                points -= 1

        return points
    
#FUNCTION FOR READING THE POINTS
def punct2(points):
    if points != 1:
        print('You have: {} points!'.format(points),'\n')
    else:
        print('You have {} point'.format(points),'\n')

#FUNCTION FOR READING THE POINTS AT THE END OF THE GAME
def final(points):
    if points <= 0 :
        print('You really have no idea about the 2020 Top tracks! :(')
    elif points >1 and points < 7:
        print('Not bad! You know some stuff about the 2020 Top tracks!')
    else:
        print('You are pro! Looking forward to the 2021 tracks?')
            
```

Here below you will find the functions for each question. Each of them creates a dictionary depending on what the question is and generates a question. To create the distractions, we use the dictionary and the function random_choice created in the cell above. Then, to add punctuation and a message with the correct answer we use the functions punct and punct 2. 

There are six questions in total. Four of them will be asked twice, meanwhile the two final, since the answer is always the same, will appear only once. Take into account that there is a question using the pictures, so a window will pop up with the image in question, you will just need to take a look at it an close it to keep on playing. 

The maximum punctuation you can obtain is 10 points. However, each time that you fail a question, 1 point will be substracted, which means that you can end up with -10 points by the end of the game! 

You can write the answer in capitals or lower-case, the game is case insensitive. Moreover, if you misspell the answer just by one letter, the programme will take it as correct (Levensthein distance) :) 

ENJOY!

```python
def questions_songs(points):
    
    #CREATING A DICTIONARY WITH THE COLUMNS WE WILL USE
    dic = dict(zip(df['artist'],df['track_name']))
    
    artist = random_choice(dic)
    q = 'Which song is sung by {}?'.format(artist)
    print(q)
    
    #CREATING DISTRACTIONS
    distractions= [dic[artist]]
    correct = dic[artist]
    while len(distractions) < 4:
        song = random.choice(df['track_name'])
        if song not in distractions:
            distractions.append(song)
    print(sorted(distractions))
    
    #ASKING FOR INPUT AND CALLING FUNCTIONS FOR PUNCTUATIONS
    answer = input()   
    points = punct(correct,answer,points)
    punct2(points)
    return points
```

```python
def questions_singers(points):
    
        #CREATING A DICTIONARY WITH THE COLUMNS WE WILL USE
        dic = dict(zip(df['artist'],df['track_name']))
        
        artist = random_choice(dic)
        song = dic[artist]
        q = 'Who is the singer of {}?'.format(song)
        print(q)
        
        #CREATING DISTRACTIONS
        distractions = [artist]
        correct = artist
        while len(distractions) < 4:
            artist = random_choice(dic)
            if artist not in distractions:
                distractions.append(artist)
        print(sorted(distractions))
        
        #ASKING FOR INPUT AND CALLING FUNCTIONS FOR PUNCTUATIONS
        answer = input()
        points = punct(correct,answer,points)
        punct2(points)
        return points
```

```python
def questions_album(points):
    
    #CREATING A DICTIONARY WITH THE COLUMNS WE WILL USE
    dic = dict(zip(df['album'],df['track_name']))
    
    album = random_choice(dic)
    song = dic[album]
    q = 'In which album can you find the song {}?'.format(song)
    print(q)
    
    #CREATING DISTRACTIONS
    distractions = [album]
    correct = album
    while len(distractions) < 4:
        album = random_choice(dic)
        if album not in distractions:
            distractions.append(album)
    print(sorted(distractions))

    #ASKING FOR INPUT AND CALLING FUNCTIONS FOR PUNCTUATIONS
    answer = input()
    points = punct(correct,answer,points)
    punct2(points)
    return points
```

```python
import webbrowser
def questions_picture(points):
    
    #CREATING A DICTIONARY WITH THE COLUMNS WE WILL USE
    photos = df[df['image'].isna() == False] #Data set containing only artists with a photo
    dic = dict(zip(photos['artist'],photos['image']))

    artist = random_choice(dic)
    picture = dic[artist]
    q = 'Who is the singer that appears in the picture?'
    print(q)

    #CREATING DISTRACTIONS
    distractions = [artist]
    correct = artist
    while len(distractions) < 4:
        artist = random_choice(dic)
        if artist not in distractions:
            distractions.append(artist)
    print(sorted(distractions))
    
    print('Type y to see the image')
    ans = input()
    if ans.lower() == 'y':
        webbrowser.open(picture)  #OPEN PICTURE IN ANOTHER WINDOW
    

    #ASKING FOR INPUT AND CALLING THE FUNCTIONS FOR PUNCTUATION
    answer = input()
    points = punct(correct,answer,points)
    punct2(points)
    return points
```

```python
def questions_artist(points):
    
    #CREATING A DICTIONARY WITH THE COLUMNS WE WILL USE
    dic = df['artist'].value_counts().to_dict()
    q = 'Who has more songs in the Top 2020 playlist?'
    print(q)
    
    #CREATING DISTRACTIONS
    distractions = [list(dic.keys())[0]]
    correct = distractions[0]
    while len(distractions) < 4:
        artist = random_choice(dic)
        if (artist not in distractions) and (artist != 'Travis Scott') and (artist != 'Billie Eilish'): #This three artist have the same number of hits (3 each).
            distractions.append(artist)
    print(sorted(distractions))
    
    #ASKING FOR INPUT AND CALLING THE FUNCTIONS FOR PUNCTUATION
    answer = input()
    points = punct(correct,answer,points)
    punct2(points)
    return points  
```

```python
def questions_top(points):
    
    #NO NEED FOR DICTIONARY THIS TIME
    q = 'Which song was at the top 1?'
    print(q)
    
    #CREATING DISTRACTIONS
    distractions = [df.iloc[0,0]]
    correct = distractions[0]
    while len(distractions) < 4:
        song = random.choice(df['track_name'])
        if song not in distractions:
            distractions.append(song)
    print(sorted(distractions))
    
    #ASKING FOR INPUT AND CALLING THE FUNCTIONS FOR PUNCTUATION
    answer = input()
    points = punct(correct,answer,points)
    punct2(points)
    return points
```

```python
print('Do you wish to start the game?\nYES!','NO!')
ans = input()
if ans.lower() == 'YES!'.lower():
    
    print('LET THE GAME ABOUT THE 2020 TOP TRACKS BEGIN!\n')
    points = 0
    print('QUESTION NUMBER 1:')
    points= questions_songs(points)
    print('QUESTION NUMBER 2:')
    points= questions_singers(points)
    print('QUESTION NUMBER 3:')
    points= questions_album(points)
    print('QUESTION NUMBER 4:')
    points= questions_picture(points)
    print('QUESTION NUMBER 5:')
    points= questions_songs(points)
    print('QUESTION NUMBER 6:')
    points= questions_singers(points)
    print('QUESTION NUMBER 7:')
    points= questions_album(points)
    print('QUESTION NUMBER 8:')
    points= questions_picture(points)
    print('QUESTION NUMBER 9:')
    points = questions_artist(points)
    print('FINAL QUESTION!')
    points = questions_top(points)
    
    print('Your final score is:...\n {} POINTS!'.format(points))
    final(points)
else:
    print('Ok, bye! :(')


```
